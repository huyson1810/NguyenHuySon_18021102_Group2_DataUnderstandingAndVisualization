```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt     
import matplotlib.ticker as ticker
```

```python
import seaborn as sns
```

```python
file_inp = pd.read_json('sosanhgia20200803_023228.txt', lines = True)
file_inp
```

```python
list_categ = file_inp['category']
list_categ
```

```python
list_cate = list(list_categ)
specific_categ = []
for items in list_cate:
    specific_categ.append(items[0])
specific_categ
```

```python
df = file_inp
df
```

```python
df.insert(2,'specific_categ',specific_categ)
df
```

```python
df.info()
```

```python
#filter data , remove data containing NaN type
df = df.dropna()
df
```

```python
list_price = df['priority_price']
list_price
```

```python
price = []
for items in list_price:
    price.append(int(items.replace('.','')))
```

```python
print(price)
```

```python
#insert 'Price Value' column to dataframe
df.insert(8,'Price Value',price)
df
```

```python
plt.rcParams['figure.figsize'] = (15,12)

sns.jointplot(x="num_of_product", y="Price Value", data=df)

plt.suptitle('The Relationship between Price and The Number of Product Respectively',y=1 ,fontsize =13,color='blue')

plt.show()
```

```python
group_cate = df.groupby('specific_categ')
```

```python
sum_gr_cate = group_cate.sum()
sum_gr_cate
```

```python
sum_gr_cate.index
```

```python
plt.rcParams['figure.figsize'] = (15,10)
x1 = sum_gr_cate.index
y1 = sum_gr_cate['num_of_store']  
y2 = sum_gr_cate['num_of_product']  + sum_gr_cate['num_of_store']

plt.xticks(fontsize =12)
plt.yticks(fontsize =14)
#plt.xticks(rotation=70)

plt.ylabel('Category',fontsize =13)
plt.xlabel('Quantity of Product',fontsize = 13)
plt.suptitle('The Statistics of Quantity of Product',fontsize = 16)

plt.barh(x1,y2,height=0.5,label = 'Total',color = 'orange')
plt.barh(x1,y1,height=0.5,label = 'InStore',color ='green')


plt.legend(shadow = True,fontsize =15)
plt.show()
```

```python
sum_gr_cate
```

```python
plt.rcParams['figure.figsize'] = (17,10)
x1 = sum_gr_cate.index
y1 = sum_gr_cate['num_of_product']
total = sum_gr_cate['num_of_product'] + sum_gr_cate['num_of_store']
y2 = ( sum_gr_cate['num_of_product'] / total ) *100

plt.xticks(rotation=70)

plt.xlabel('Category',fontsize =13)
plt.ylabel('Proportion (%)',fontsize =13)
plt.suptitle('The Statistics of Consuming Product Rate',fontsize =16)

plt.bar(x1,y2,color = 'orange',width=0.5, label = 'Consumption')
#plt.barh(x1,y2,height=0.5,label = 'Total')


plt.legend()
plt.show()
```

```python
group_brand = group_cate.get_group('Máy tính - Thiết bị VP').groupby('brand')
```

```python
computer_mean = group_brand.mean()
computer_mean
```

```python
computer_max = group_brand.max()
computer_max
```

```python
plt.rcParams['figure.figsize'] = (15,12)
x1 = computer_mean.index
y1 = computer_mean['Price Value'] / 10**6
y2 = computer_max['Price Value'] / 10**6

plt.xticks(rotation=30)
plt.grid()
plt.xlabel('Brand',fontsize =13)
plt.ylabel('Million VND ',fontsize =13)
plt.suptitle('The Selling Price of Available Computer Brands',fontsize =16)

plt.plot(x1, y2, label='Max Price',color ='red')
plt.bar(x1,y1,width=0.5,label = 'Average Price')

plt.legend()
plt.show()
```

```python
gr_sach = group_cate.get_group('Sách')
gr_sach
```

```python
gr_sach_store = gr_sach.groupby('priority_store')
```

```python
sum_sach_store = gr_sach_store.sum()
sum_sach_store
```

```python
store = sum_sach_store.index
store
```

```python
sum_all = 0
for i in sum_sach_store['num_of_store']:
    sum_all += i
sum_all    
```

```python
rate_sach_store = sum_sach_store['num_of_store'] / sum_all * 100
rate_sach_store
```

```python
plt.rcParams['figure.figsize'] = (20,12)
Explode = [0,0,0,0.02,0,0,0.02,0.05,0]
plt.pie(rate_sach_store, explode=Explode, labels=store , shadow =False , startangle=90)
plt.axis('equal')
plt.suptitle('The Proportion of Available Book in Stores',fontsize =20)
plt.legend(title = "Store' Name",fontsize = 14, shadow = True)
plt.show()
```

```python
consume = gr_sach['num_of_product']
in_store = gr_sach['num_of_store']
```

```python
colors = consume / in_store
colors
```

```python
plt.rcParams['figure.figsize'] = (20,12)

plt.scatter(consume,in_store , s=200 , c=colors  , cmap = 'Reds' 
            ,edgecolor = 'black' , linewidths=.5 )
plt.grid()

plt.title("Scatter The Ratio between Books Sold and Books In Store",fontsize=16)

plt.xlabel('Consumption',fontsize=14)
plt.ylabel('In Store',fontsize=14)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.colorbar()

plt.show()
```

```python
from mpl_toolkits import mplot3d
```

```python
ax = plt.axes(projection = '3d')
ax.scatter3D(consume,in_store, s=100 , c=colors  , cmap = 'Reds' 
            ,edgecolor = 'black' , linewidths=.5)
ax.set_xlabel('Consumption',fontsize=14)
ax.set_ylabel('In Store',fontsize=14)

plt.title("Scatter 3D The Ratio between Books Sold and Books In Store",fontsize=16)

plt.show()
```
